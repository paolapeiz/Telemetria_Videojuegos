```python
import pandas as pd

df = pd.read_csv('videojuegos_online_telemetria.csv')
display(df.head())
```

```python
print(f"Dimensiones del DataFrame antes de eliminar duplicados por PlayerID: {df.shape}")

df.drop_duplicates(subset=['PlayerID'], inplace=True)

print(f"Dimensiones del DataFrame después de eliminar duplicados por PlayerID: {df.shape}")
```

```python
min_age = df['Age'].min()
max_age = df['Age'].max()

print(f"Edad mínima de los jugadores: {min_age}")
print(f"Edad máxima de los jugadores: {max_age}")
```

```python
# Reemplazar 'Europe' por 'EU'
df['Region'] = df['Region'].replace('Europe', 'EU')

# Rellenar valores NaN con 'EEUU'
df['Region'] = df['Region'].fillna('EEUU')

print("Valores distintos y su conteo en la columna 'Region' después de las transformaciones:")
display(df['Region'].value_counts(dropna=False))
```

```python
print("Valores nulos por columna:")
display(df.isnull().sum())
```

### Estandarización de mayúsculas/minúsculas en columnas de texto

Vamos a estandarizar las columnas de tipo 'object' (texto) convirtiendo todos sus valores a minúsculas. Esto unificará entradas que son iguales pero tienen diferentes capitalizaciones.

```python
object_cols = df.select_dtypes(include='object').columns

print("Columnas de tipo 'object' a estandarizar:", object_cols.tolist())

for col in object_cols:
    # Solo aplicar si la columna no es 'PlayerID', ya que los IDs suelen ser case-sensitive y no necesitan estandarización de casing
    if col != 'PlayerID':
        if df[col].apply(lambda x: isinstance(x, str)).any(): # Ensure there are string values to process
            print(f"\nValores únicos en la columna '{col}' ANTES de la estandarización:")
            display(df[col].value_counts(dropna=False))
            df[col] = df[col].astype(str).str.lower()
            print(f"Valores únicos en la columna '{col}' DESPUÉS de la estandarización:")
            display(df[col].value_counts(dropna=False))
```

```python
columns_to_impute = ['AvgSessionMinutes', 'WinRate', 'InGamePurchase', 'FriendsInGame']

for col in columns_to_impute:
    if df[col].isnull().any():
        mean_value = df[col].mean()
        df[col] = df[col].fillna(mean_value)
        print(f"Columna '{col}': valores nulos rellenados con la media ({mean_value:.2f})")

print("\nValores nulos por columna después de la imputación:")
display(df.isnull().sum())
```
